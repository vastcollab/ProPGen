```python
import scanpy as sc

# specify the path to .h5ad file
file_path = "/Users/asapp/Downloads/source_data/fig5_ab_geno.h5ad"

# load the AnnData object
adata = sc.read_h5ad(file_path)
```

```python
adata
```

```python
# leiden is phenotype; one hot encoding of genotype (final 3 cols)
adata.obs.head(15)
```

```python
# I think experiment is 1 of the 4 time course points they sampled

unique_experiments = adata.obs['experiment'].nunique()
print(unique_experiments)

```

```python
import seaborn as sns

# Count the frequency of each leiden number for each time point
leiden_frequency = adata.obs.groupby(['experiment', 'leiden']).size().reset_index(name='frequency')

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(data=leiden_frequency, x='experiment', y='frequency', hue='leiden')
plt.title('Leiden Number Frequency vs. Time Points')
plt.xlabel('Time Points')
plt.ylabel('Frequency')
plt.legend(title='Leiden Number')
plt.show()
```

```python
# Count the frequency of each leiden number for each time point
leiden_frequency = adata.obs.groupby(['experiment', 'leiden']).size().reset_index(name='frequency')

max_frequency = leiden_frequency['frequency'].max()
leiden_frequency['normalized_frequency'] = leiden_frequency['frequency'] / max_frequency

# Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=leiden_frequency, x='experiment', y='normalized_frequency', hue='leiden', marker='o')
plt.title('Leiden Number Frequency vs. Time Points (Normalized)')
plt.xlabel('Time Points')
plt.ylabel('Normalized Frequency')
plt.legend(title='Leiden Number')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.fishplot(data=leiden_frequency, x='experiment', y='normalized_frequency', hue='leiden')
plt.title('Leiden Number Frequency vs. Time Points (Normalized) - Fishplot')
plt.xlabel('Time Points')
plt.ylabel('Normalized Frequency')
plt.legend(title='Leiden Number')
plt.show()
```
